# Mini Project #1 : Data Analysis on Covid-19 dataset
**Tools Used**: Jupyter Notebook<br>
**Dataset**: 'country_wise_latest.csv'<br>
**Goal**: Perform exploratory data analysis (EDA) on COVID-19 data to identify trends and patterns across countries.

## Importing libraries and dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
df = pd.read_csv('country_wise_latest.csv')
```

## Basic Data Exploration

```python
df.head()
df.tail()
df.shape
df.info()
```

## Visualizations

### Confirmed Cases vs Deaths

```python


df.plot(kind = 'scatter', x = 'Confirmed', y = 'Deaths')
plt.title('Confirmed Cases vs Deaths')

plt.show()
```

### Histogram (Confirmed cases)

```python
df['Confirmed'].plot(kind = 'hist')
plt.title('Confirmed cases')
plt.xlabel('Confirmed cases')

plt.show()
```

## Bar chart (Top 10 countries by confirmed cases)

```python
plt.figure(figsize=(10, 6))
top_confirmed = df.sort_values(by = 'Confirmed', ascending = False).head(10)

plt.bar(top_confirmed['Country/Region'], top_confirmed['Confirmed'])

plt.xlabel('Country')
plt.ylabel('Confirmed')
plt.title('Top 10 countries by confirmed cases')

plt.show()
```

## Correlation Heatmap

```python
numeric_columns = ['Confirmed', 'Deaths', 'Recovered', 'Active']

correlation = df[numeric_columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot= True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation between COVID-19 metrics')
plt.show()
```

## Conclusion
- Confirmed cases and deaths are positively correlated across countries.
- Most countries have relatively low confirmed cases, with a few showing very high numbers.
- The top 10 countries by confirmed cases dominate the global figures.
- Correlation heatmap highlights strong relationships between key COVID-19 metrics.
